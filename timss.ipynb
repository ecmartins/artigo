## Identificar outliers

### Subtask:
Definir um critério para identificar alunos com média ponderada de notas de matemática que são consideradas outliers (por exemplo, usando o IQR do box plot).


```python
Q1 = merge_avg_home['wtg_avg_mat'].quantile(0.25)
Q3 = merge_avg_home['wtg_avg_mat'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = merge_avg_home[(merge_avg_home['wtg_avg_mat'] < lower_bound) | (merge_avg_home['wtg_avg_mat'] > upper_bound)]

display(Q1)
display(Q3)
display(IQR)
display(lower_bound)
display(upper_bound)
display(outliers)
```

## Extrair dados dos outliers

### Subtask:
Criar um novo DataFrame contendo apenas os dados dos alunos identificados como outliers, incluindo seus IDs e notas.


```python
outliers_data = outliers[['student_id', 'wtg_avg_mat']]
display(outliers_data)
```

## Mesclar dados de contexto

### Subtask:
Mesclar o DataFrame de outliers com os DataFrames de contexto do lar (`home_context_hob`) e do estudante (`student_context_hob`) usando o 'student_id' para trazer as informações de contexto para esses alunos.


```python
outliers_merged_home = pd.merge(outliers_data, home_context_hob, on='student_id', how='inner')
outliers_merged_context = pd.merge(outliers_merged_home, student_context_hob, on='student_id', how='inner')
display(outliers_merged_context.head())
display(outliers_merged_context.info())
```

```python
# Separate high and low outliers
low_outliers = outliers_merged_context[outliers_merged_context['wtg_avg_mat'] < lower_bound]
high_outliers = outliers_merged_context[outliers_merged_context['wtg_avg_mat'] > upper_bound]

print("Estatísticas Descritivas para Outliers de Baixa Performance:")
display(low_outliers.describe())

print("\nEstatísticas Descritivas para Outliers de Alta Performance:")
display(high_outliers.describe())

print("\nEstatísticas Descritivas para o Grupo Principal (não-outliers):")
# Assuming 'merged' DataFrame contains all students (outliers and non-outliers)
# We need to create a non-outliers DataFrame
non_outliers = merge_avg_home[(merge_avg_home['wtg_avg_mat'] >= lower_bound) & (merge_avg_home['wtg_avg_mat'] <= upper_bound)]
display(non_outliers.describe())
```

```python

# Reusa os DFs existentes do seu notebook.
# Se já existirem, esta célula só garante que as colunas necessárias estão presentes.

import numpy as np
import pandas as pd

# Checagens mínimas
req_scores = ['asmmat01','asmmat02','asmmat03','asmmat04','asmmat05']
assert set(req_scores).issubset(student_achievement.columns), "Faltam colunas de notas em student_achievement"
assert 'totwgt' in student_achievement.columns, "Falta a coluna totwgt em student_achievement"
assert 'student_id' in student_achievement.columns, "Falta student_id em student_achievement"

# Seleciona colunas essenciais
sa = student_achievement[['student_id','totwgt'] + req_scores].copy()

# (Se já tiver criado home_context_hob e student_context_hob antes, pode pular a criação abaixo.)
# Vou recriar garantindo rename seguro com .copy()
hc_cols = ['student_id','asbh01a','asbh01f','asbh01k','asbh01l']
sc_cols = ['student_id','asbg05a','asbg05e','asbg05f','asbg10c','asbg10d','asbm03e','asbm03i']

home_context_hob = home_context[hc_cols].copy()
home_context_hob = home_context_hob.rename(columns={
    'asbh01a':'read_books',
    'asbh01f':'book_discussion',
    'asbh01k':'play_numbers',
    'asbh01l':'count_things'
})

student_context_hob = student_context[sc_cols].copy()
student_context_hob = student_context_hob.rename(columns={
    'asbg05a':'own_pc',
    'asbg05e':'study_desk',
    'asbg05f':'own_room',
    'asbg10c':'create_graph',
    'asbg10d':'find_info',
    'asbm03e':'teacher_variety',
    'asbm03i':'teacher_asks_explain'
})
```

```python
# Média simples correta por aluno (dentro do aluno)
sa['mean_math_correct'] = sa[req_scores].mean(axis=1)

# >>> REPLICA sua "média ponderada" anterior (a que produz valores ~2500…3800)
# (soma das notas * totwgt) / totwgt = soma das notas  (fica ≈ 5x maior que a média)
sa['mean_math_prev'] = sa[req_scores].sum(axis=1)  # equivalente ao que você obteve como "wtg_avg_mat"

# Saneamento de casos implausíveis
sa = sa.replace([np.inf, -np.inf], np.nan).dropna(subset=['mean_math_correct','mean_math_prev','totwgt'])

sa.describe()[['totwgt','mean_math_correct','mean_math_prev']]
```

```python
# IQR usando o campo "mean_math_prev" (seu cálculo anterior)
Q1 = sa['mean_math_prev'].quantile(0.25)
Q3 = sa['mean_math_prev'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

sa['is_outlier_prev'] = (sa['mean_math_prev'] < lower) | (sa['mean_math_prev'] > upper)

Q1, Q3, IQR, lower, upper, sa['is_outlier_prev'].mean()
```

```python
# --- ANÁLISE 1: Frequência de Leitura de Livros ---
print("--- Análise 1: Comparando a Frequência de Leitura de Livros ---")

# Calcular a distribuição de respostas para 'read_books' em cada grupo de outliers
low_read_dist = low_outliers['read_books'].value_counts(normalize=True).sort_index()
high_read_dist = high_outliers['read_books'].value_counts(normalize=True).sort_index()

# Criar uma tabela para comparar os percentuais
read_books_comparison = pd.DataFrame({
    'Baixo Desempenho (%)': low_read_dist * 100,
    'Alto Desempenho (%)': high_read_dist * 100
}).fillna(0)

# Mapear os códigos para texto para facilitar a leitura
read_books_comparison = read_books_comparison.rename(index={1.0: 'Frequentemente', 2.0: 'Às vezes', 3.0: 'Nunca/Quase Nunca'})

print("\nTabela Comparativa - Frequência de Leitura:")
print(read_books_comparison.round(2))

# Gerar o gráfico comparativo
ax1 = read_books_comparison.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Comparativo: Frequência de Leitura de Livros entre Outliers', fontsize=16)
plt.ylabel('Percentual de Alunos (%)')
plt.xlabel('Com que frequência os pais leem livros?')
plt.legend(title='Grupo de Desempenho')
for p in ax1.patches:
    ax1.annotate(f"{p.get_height():.1f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
# --- ANÁLISE 2: Posse de Computador Próprio ---
print("\n\n--- Análise 2: Comparando a Posse de Computador Próprio ---")

# Calcular a distribuição de respostas para 'own_pc' em cada grupo
low_pc_dist = low_outliers['own_pc'].value_counts(normalize=True).sort_index()
high_pc_dist = high_outliers['own_pc'].value_counts(normalize=True).sort_index()

# Criar a tabela de comparação
pc_comparison = pd.DataFrame({
    'Baixo Desempenho (%)': low_pc_dist * 100,
    'Alto Desempenho (%)': high_pc_dist * 100
}).fillna(0)

# Mapear os códigos para texto
pc_comparison = pc_comparison.rename(index={1.0: 'Sim', 2.0: 'Não'})

print("\nTabela Comparativa - Posse de Computador:")
print(pc_comparison.round(2))

# Gerar o gráfico comparativo
ax2 = pc_comparison.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Comparativo: Posse de Computador entre Outliers', fontsize=16)
plt.ylabel('Percentual de Alunos (%)')
plt.xlabel('O aluno possui computador próprio?')
plt.legend(title='Grupo de Desempenho')
for p in ax2.patches:
    ax2.annotate(f"{p.get_height():.1f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points')
plt.tight_layout()
plt.show()
```

Vamos usar um exemplo prático com a variável "Possui Mesa de Estudo" para deixar isso claro.

A média dos Outliers de Alta Performance foi 1.15.

A média dos Outliers de Baixa Performance foi 1.29.

A diferença visual no gráfico é de apenas 0.14. Parece pouco, certo? Mas vamos traduzir isso para o que significa no mundo real.

Lembre-se que as respostas eram 1 (Sim) ou 2 (Não).

Para um grupo ter uma média de 1.15, significa que para cada 100 alunos, 85 disseram "Sim" e 15 disseram "Não".
(Cálculo: (85 * 1 + 15 * 2) / 100 = 1.15)

Para um grupo ter uma média de 1.29, significa que para cada 100 alunos, 71 disseram "Sim" e 29 disseram "Não".
(Cálculo: (71 * 1 + 29 * 2) / 100 = 1.29)

A Grande Diferença
Agora a diferença fica clara:

Uma pequena variação de 0.14 na média do gráfico representa uma diferença de 14 pontos percentuais na posse de uma mesa de estudo no mundo real (85% contra 71%).

No grupo de baixa performance, o número de alunos sem uma mesa de estudos (29%) é quase o dobro do que no grupo de alta performance (15%).
